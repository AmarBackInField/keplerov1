```python
from livekit import api
```

```python
from dotenv import load_dotenv
import os
load_dotenv()
# ------------------------------------------------------------
# Environment / LiveKit admin credentials (fetched from env)
# ------------------------------------------------------------
LIVEKIT_URL = os.getenv("LIVEKIT_URL")
LIVEKIT_API_KEY = os.getenv("LIVEKIT_API_KEY")
LIVEKIT_API_SECRET = os.getenv("LIVEKIT_API_SECRET")
```

```python
lkapi = api.LiveKitAPI(api_key=LIVEKIT_API_KEY, api_secret=LIVEKIT_API_SECRET, url=LIVEKIT_URL)
room_service = lkapi.room
```

```python
room_service.delete_room('my-assistant-room');
```

```python
active_rooms = await room_service.delete_room(api.ListRoomsRequest())
```

```python
from livekit import api

# First, list all active rooms to see what rooms exist
list_request = api.ListRoomsRequest()
active_rooms = await room_service.list_rooms(list_request)

print("Active rooms:")
for room in active_rooms.rooms:
    print(f"- {room.name} (SID: {room.sid})")

# Then delete a specific room by name
delete_request = api.DeleteRoomRequest(room="your-room-name-here")
await room_service.delete_room(delete_request)
```
